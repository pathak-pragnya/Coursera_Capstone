# Importing library

```python
import pandas as pd
```

# Importing Dataset 

```python
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
df = pd.read_html(url)[0]
df.head()
```

# Cleaning Datasets

```python
df.drop(df.loc[df['Borough']=='Not assigned'].index, inplace=True)
df.reset_index(drop=True,inplace=True)
df.head()
```

```python
df.replace(df.loc[(df['Neighborhood'] == 'Not assigned') & (df['Borough']!='Not assigned')],df['Borough'],inplace=True)
df = df.groupby(['Postal Code','Borough'])['Neighborhood'].apply(','.join).reset_index()
df.head()
```

```python
df.shape
```

# Getting Coordinates

```python
!wget -O geospatial.csv https://cocl.us/Geospatial_data
df_latlng = pd.read_csv('geospatial.csv')
```

```python
final_df = pd.merge(df, df_latlng, on=['Postal Code'])
```
